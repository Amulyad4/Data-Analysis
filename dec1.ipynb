## columns and their type
Loan_ID	-- nominal -- not in a order 

Gender -- nominal -- no order/ranking

Married	-- nominal -- no order/ranking

Dependents -- discrete -- distinct numeric values

Education -- nominal -- no order/ranking

Self_Employed -- nominal -- no order/ranking

ApplicantIncome	-- continuous -- large range of value

CoapplicantIncome -- continuous -- large range of value

LoanAmount -- discrete -- multiples of 12 

Loan_Amount_Term -- continuous -- large range of value

Credit_History -- nominal -- no order/ranking

Property_Area -- ordinal -- in order/ranking

Loan_Status (Approved) -- nominal -- no order/ranking

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
df['Gender'].unique()
```

```python
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
df['Married'].unique()
```

```python
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Dependents'].unique()
```

```python
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['Self_Employed'].unique()
```

```python
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].median())
df['LoanAmount'].unique()
```

```python
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median())
df['Loan_Amount_Term'].unique()
```

```python
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mode()[0])
df['Credit_History'].unique()
```
