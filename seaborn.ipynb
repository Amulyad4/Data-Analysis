# Seaborn - Statistical Data Visualization

## 1. Why Learn Seaborn?
#### What is it ?
Seaborn is a python library that builds on the top of matplaotlib, that makes the statistical data more easy and beautiful. 

#### Why we need it ?
1. For creating plots like violin plot, box plot and pair plots.

2. Automatically handles data from pandas data frames.

#### Where is it used?
1. Data Analysis and reporting

2. Exploring relationships between variables

3. Machine Learning for visualizing data sets and results

## 2. Import Seaborn

```python
import seaborn as sns
```

## 3. Load Dataset

```python
# Load Titanic dataset
df = sns.load_dataset("titanic")
df.head()
 
```

## 4. Histograms & Countplots


```python
import matplotlib.pyplot as plt 

# Histogram
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

## 5. Scatter Plots

```python

sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

## 6. Boxplot & Violin Plot

```python

# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

## 7. Pairplot

```python

iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

## 8. Heatmap

```python

# Correlation heatmap
corr = iris.corr()
# corr = iris.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

## 9. Customization & Themes

```python

sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
 
 
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
 
```

```python
titanic = sns.load_dataset('titanic')
```

```python
# 1.Compare Passenger Survival by Class and Gender
plt.figure(figsize=(7,5))
sns.barplot(data=titanic,x='class', y='survived',hue='sex',palette='Set1')
plt.title("Passenger Survival by Class and Gender")
plt.xlabel("Passenger Class")
plt.ylabel("Survival rate")
plt.show()
```

```python
# 2. Distribution of Passenger Ages
plt.figure(figsize=(7,5))
sns.histplot(titanic['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# 3. Fare Distribution by Class
plt.figure(figsize=(7,5))
sns.boxplot(x='class', y='fare', data=titanic, palette='pastel')
plt.title("Fare Distribution by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python

# 4. Relationship Between Age and Fare
plt.figure(figsize=(7,5))
sns.scatterplot(x='age', y='fare', hue='class', data=titanic, alpha=0.7)
plt.title("Relationship Between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
# 5. Correlation Heatmap
plt.figure(figsize=(7,5))
corr = titanic[['survived', 'pclass', 'age', 'fare', 'sibsp', 'parch']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Titanic Features")
plt.show()
```
